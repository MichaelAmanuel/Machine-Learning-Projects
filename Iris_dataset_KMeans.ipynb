```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Generate random data
np.random.seed(42)  # For reproducibility
X = np.random.rand(100, 2)  # 100 points in 2D space
```

```python
X
```

```python
# Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
```

```python
# Get cluster assignments and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
# Plot the data points and color them by cluster
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o')

# Plot the centroids
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', marker='x')

plt.title('KMeans Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
# Step 1: Generate random data
np.random.seed(42)  # For reproducibility
X = np.random.rand(100, 2)  # 100 points in 2D space

# Step 2: Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Step 3: Get cluster assignments and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Step 4: Visualize the results
# Plot the data points and color them by cluster
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o')

# Plot the centroids
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', marker='x')

plt.title('KMeans Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
# Load Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Get cluster assignments and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Reduce dimensionality to 2D using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot the data points and color them by cluster
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', marker='o')

# Plot the centroids (also reduced to 2D)
centroids_pca = pca.transform(centroids)
plt.scatter(centroids_pca[:, 0], centroids_pca[:, 1], s=300, c='red', marker='x')

plt.title('KMeans Clustering on Iris Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
